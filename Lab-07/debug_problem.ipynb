Python import statements 

```python
import PIL.Image as image 
import numpy as np 
import matplotlib.pyplot as plt 
```

Load a image and show the orginal image

```python
img_array = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90],
])

plt.imshow(img_array, cmap="gray")
plt.title("Orginal Image")
plt.axis("off")

plt.show()
```

Image Height and Width using shape

```python
H, W = img_array.shape
print("Height * Width : ", H, "*",  W)
```

filtering metrix

```python
filter = np.array([
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
    [1/9, 1/9, 1/9],
])
print("Filter: ", filter)
```

replicate padding

```python
padded_img = np.pad(img_array, pad_width=1, mode="edge")
print("Padding Image: ", padded_img)
```

Mean filterig method

```python
mean_result = np.zeros_like(img_array)

for i in range(H):
    for j in range(W):
        region = padded_img[i : i + 3, j : j + 3]
        mean_result[i, j] = np.sum(region * filter)

print("Mean filter: ", mean_result)
```

Median filtering method

```python
median_result = np.zeros_like(img_array)

for i in range(H):
    for j in range(W):
        region = padded_img[i : i + 3, j : j + 3]
        median_result[i, j] = np.median(region)

print("Median filter: ", median_result)
```

Display both image 

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.imshow(img_array, cmap="gray")
plt.title("Orginal Image")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(mean_result, cmap="gray")
plt.title("Mean filter Image")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(median_result, cmap="gray")
plt.title("Median filter Image")
plt.axis("off")
```
